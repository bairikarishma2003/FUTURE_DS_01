```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('../data/sales_data.csv')

# Show first 5 rows
df.head()
```

```python
# Convert Date column
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Create Revenue column
df['Revenue'] = df['Quantity'] * df['Price']

# Show updated data
df.head()
```

```python
print("Total Revenue:", df['Revenue'].sum())
```

```python
top_products = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
top_products
```

```python
top_products.plot(kind='bar')
plt.title("Top Selling Products")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

```python
region_sales = df.groupby('Region')['Revenue'].sum()

region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales by Region")
plt.show()
```

```python
category_sales = df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)
category_sales
```

```python
category_sales.plot(kind='bar')
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python
df.groupby('Date')['Revenue'].sum().plot(kind='line')
plt.title("Revenue Trend Over Time")
plt.show()
```

```python
df.groupby('Category')['Revenue'].sum().plot(kind='bar')
plt.title("Revenue by Category")
plt.show()
```

```python
df.groupby('Region')['Revenue'].sum().plot(kind='bar')
plt.title("Revenue by Region")
plt.show()
```

📊 Insights
The highest revenue is generated by Laptop, making it the top selling product.
Electronics category dominates overall sales compared to Clothing and Accessories.
North region contributes the highest revenue among all regions.
Accessories and Footwear generate comparatively lower revenue.
Business should focus more on high performing products like Laptop and Phone.
